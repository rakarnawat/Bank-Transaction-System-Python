```python
#t1,t2=try,transc try
#tran=transc type 
#a=random no
#otp=otp
import openpyxl 
from openpyxl import load_workbook
import pandas as pd # import xlsr for excel files
xls=pd.ExcelFile("F:/project/SITIP19-KIIT-Python-FinalProject1-Workbook.xlsx")
df=pd.read_excel(xls,"Sheet1")

import numpy as np
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

l=['Sr. No.', 'User ID', 'A/c No.', 'Full Name', 'User Email','A/c Balance', 'A/c Status']
for i in df:
    if i not in l:
        df.drop(i,axis=1,inplace=True)

block='false'
#transaction function-------------------------------------------------------------------------------------------------------
def trans():
    for i in range(3):
        t2=input("Do you want to do the transaction (y/n) : ").lower()
        if t2=='y':
            tran=input("Which type of transaction you want to do : \n1)‘C’ for credit money \n2)‘D’ for debit money \nEnter :").lower()
        elif t2=='n':
            break
        else:
            print("Not a valid option !\n")
            break
    
        if tran=='c':
            print("Your Current Balance is : "+ str(df["A/c Balance"][index])+"rs\n")
            credit=int(input("Enter the amount You want to credit : "))
            print("Successful Transaction of Rs.%i completed "%(credit))
            #credit+=balance
            df["A/c Balance"][index]+=credit
            print("Available balance is : " + str(df["A/c Balance"][index]) +"rs\n")
            break
        elif tran=='d':
            print("Your Current Balance is : "+ str(df["A/c Balance"][index])+"rs\n")
            debit=int(input("Enter the amount You want to debit : "))
            if debit<=df["A/c Balance"][index]:
                print("Successful Transaction of Rs.%i completed "%(debit))
                #debit=balance-debit
                df["A/c Balance"][index]-=debit
                print("Available balance is : " + str(df["A/c Balance"][index]) +"rs\n")
                break
            else:
                print("You Dont Have Sufficient Money To Debit ! \nMaximum Debit limit of your account is : "+str(df["A/c Balance"][index])+"rs\n")
                t1=input("Do you Want to try again (y/n) : ").lower()
                if t1=='y':
                    trans()
                elif t1=='n':
                    break
                else:
                    print("Not Valid Option !")
                    break
            
        elif i<2 and (tran!='c' or tran!='d')  :
            print("Please Enter a Valid Transaction Choice !")
            print("You have %i Attempt Left !\n"%(3-i))
        else :
            print("YOUR ACCOUNT HAS BEEN BLOCKED FOR SECURITY REASONS !\n")
            df["A/c Status"][index]="Blocked"
            block='true'
            
#user validation-----------------------------------------------------------------------------------------------------------
u=-1
for i in range(3):
    yes=2
    user=input("Enter Valid Username Or User ID : ")
    try:
        user=eval(user)
    except :
        user=user
    if type(user)==type("22"):
        u=0
        for j in range(len(df)):
            if df.loc[j,"Full Name"]==user:
                yes=1
                index=j
                break
    if type(user)==type(22):
        u=1
        for j in range(len(df)):
            if df.loc[j,"User ID"]==user:
                index=j
                yes=1
                break    
    
    if (yes==1) :
        break

    elif i<2 :
        print("Invalid Username/User ID !")
        print("You have %i Attempt Left !\n"%(2-i))
    else :
        print("You have entered invalid Username/User ID 3 times . You need to start again")
        block='true'
        break
    
        
if block=='false':
    for i in range(3):
        try:
            accn=input("Enter Valid Account Number : ")
        except:
            if i<2 :
                print("Invalid Account Number !")
                print("You have %i Attempt Left !\n"%(2-i))
                continue
            else :
                print("Entered Account Number is invalid !\n YOUR ACCOUNT HAS BEEN BLOCKED FOR SECURITY REASONS !\n")
                df["A/c Status"][index]="Blocked"
                block='true'
                print(index)
                break
        else :       
            if (accn==df["A/c No."][index]):
                break
            elif i<2 and (accn!=df["A/c No."][index]):
                print("Invalid Account Number !")
                print("You have %i Attempt Left !\n"%(2-i))
            else :
                print("Entered Account Number is invalid !\n YOUR ACCOUNT HAS BEEN BLOCKED FOR SECURITY REASONS !\n")
                df["A/c Status"][index]="Blocked"
                block='true'
                break

else:
    block='true'
            
if block=='false' and df["A/c Status"][index]=="Active":
    #OTP generation---------------------------------------------------------------------------------------------------
    a1=np.random.randint(1000,10000)
    #print(a1) 

    #OTP send to mail------------------------------------------------------------------------------------------------
    fromaddr = "YOUR_GMAIL"
    toaddr = df["User Email"][index]
    msg = MIMEMultipart()

    msg['From'] = fromaddr
    msg['To'] = toaddr
    msg['Subject'] = 'OTP Verification !'

    body = "Hi! Your 4 digit OTP is : " + str(a1)
    msg.attach(MIMEText(body,'plain'))

    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(fromaddr,"YOUR_PASSWORD")
    text = msg.as_string()
    server.sendmail(fromaddr,toaddr,text)
    server.quit()

    #OTP check (3 time check)---------------------------------------------------------------------------------------------
    if block=='false':    
        for i in range(3):
            try:
                otp=int(input('Enter The OTP sent on your given mail-id : '))
            except:
                if i<2  :
                    print("Please Enter a Valid OTP !")
                    print("You have %i Attempt Left !\n"%(2-i))
                else :
                    print("Entered OTP is invalid !\n YOUR ACCOUNT HAS BEEN BLOCKED FOR SECURITY REASONS !\n")
                    df["A/c Status"][index]="Blocked"
                    block='true'
            else:
                if otp==a1 :
                    credit=df["A/c Balance"][index]
                    trans()
                    break
        
                elif i<2 and otp!=a1 :
                    print("Please Enter a Valid OTP !")
                    print("You have %i Attempt Left !\n"%(2-i))
                else :
                    print("Entered OTP is invalid !\n YOUR ACCOUNT HAS BEEN BLOCKED FOR SECURITY REASONS !\n")
                    df["A/c Status"][index]="Blocked"
                    block='true'
        
    else:
        block='true'
    
    #Mini-Statement---------------------------------------------------------------------------------------------------------
    if block=='false':
        if credit<=df["A/c Balance"][index]:
            c="(Credit)"
            d="+"
        else:
            c="(Debit)"
            d=""
    
        print("====================================================================")
        print("|             -: Mini-Statement For Your Transaction :-            |")
        print("====================================================================")
        print("|    User Full Name               | %30s |"%(df["Full Name"][index]))
        print("|------------------------------------------------------------------|")
        print("|    User ID                      | %30s |"%(df["User ID"][index]))
        print("|------------------------------------------------------------------|")
        print("|    Registered Email-ID          | %30s |"%(df["User Email"][index]))
        print("|------------------------------------------------------------------|")
        print("|    Account Status               | %30s |"%(df["A/c Status"][index]))
        print("|------------------------------------------------------------------|")
        print("|    Previous Balance             | %30s |"%(credit))
        print("|------------------------------------------------------------------|")
        print("|    Last Transaction  %8s   | %27s%s |"%(c,d,df["A/c Balance"][index]-credit))
        print("|------------------------------------------------------------------|")
        print("|   Available Balance             | %30s |"%(df["A/c Balance"][index]))
        print("====================================================================")
    
    
    else:
        block='true'
else :
    print("Opps! You Account Is Blocked!!")
df.to_excel("F:/project/SITIP19-KIIT-Python-FinalProject1-Workbook.xlsx")    
#END--------------------------------------------------------------------------------------------------------------------
```
